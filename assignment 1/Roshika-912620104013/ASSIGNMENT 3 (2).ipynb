###Data Preprocessing

```
data.drop('id',axis=1, inplace=True)
data
```

```
data.drop('Date',axis=1, inplace=True)
data
```

```
data.dropna(inplace = True)
```

```
for x in data.index:
  if data.loc[x, "Renovation Year"] == 0:
    data.drop(x, inplace = True)
```

```
data['Renovation Year'].value_counts()
```

```
data
```

```
data['number of bathrooms'] = data['number of bathrooms'].astype(int)
```

```
data
```

```
data['number of floors'] = data['number of floors'].astype(int)
data
```

```
for x in data.index:
  if data.loc[x, "Renovation Year"] < data.loc[x,'Built Year']:
    data.drop(x, inplace = True)
```

```
data
```
